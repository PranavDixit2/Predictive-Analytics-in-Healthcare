```python
label_counts = df["diagnosis"].value_counts()
print(label_counts)
```

```python
pivoted_data = df.pivot_table(index='subject_id', columns='label', values='valuenum', aggfunc='first')


pivoted_data.columns = [f"{col[1]}_{col[0]}" for col in pivoted_data.columns]


pivoted_data.reset_index(inplace=True)


pivoted_data = pivoted_data.merge(df[['subject_id', 'diagnosis']], on='subject_id')


pivoted_data.fillna(0, inplace=True)


print(pivoted_data.sample(1))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
pivoted_data['diagnosis'] = le.fit_transform(pivoted_data['diagnosis'])

X = pivoted_data.iloc[:, 1:10]  # Features (columns 1-15)
y = pivoted_data.iloc[:, 10]    # Target variable (column 16)

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Feature Selection
# Select the best k features based on the ANOVA F-value
k_best_features = SelectKBest(score_func=f_classif, k=8)
X_new = k_best_features.fit_transform(X_scaled, y)

#
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=14)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred

from sklearn.metrics import f1_score, accuracy_score, roc_auc_score

f1 = f1_score(y_test, y_pred, average='weighted')
acc = accuracy_score(y_test, y_pred)


print(f'KNN F1: {f1}')
print(f'KNN Accuracy: {acc}')
```
